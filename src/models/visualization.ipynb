```python
import yfinance as yf
import pandas as pd
```

```python
# Fetch historical data for BTC-USD
btc_df = yf.download('BTC-USD', start='2014-01-01', end='2024-01-01').reset_index()

btc_df.head()
```

## Data Cleaning

```python
from ta import add_all_ta_features

btc_df = add_all_ta_features(btc_df, open="Open", high="High", low="Low", close="Close", volume="Volume", fillna=True)
```

```python
btc_df.head()
```

```python
btc_df = btc_df.drop(['Date', 'others_dr', 'others_dlr', 'others_cr'], axis=1)
```

```python
btc_df['next'] =  btc_df['Close'].shift(-1)
```

```python
btc_df = btc_df.dropna()
```

```python
btc_df
```

## Building Model

```python
from pycaret.regression import *
s = setup(btc_df, target = 'next', session_id = 123)
```

```python
best = compare_models()
```

```python
best = create_model('br')
final_btc_model = finalize_model(best)
```

```python
save_model(final_btc_model, model_name='eth_br_model')
```
