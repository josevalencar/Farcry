```python
import cryptocompare
import pandas as pd
```

```python
# Fetching historical daily prices for BTC in BRL
prices = cryptocompare.get_historical_price_day('BTC', currency='BRL', limit=30)
df = pd.DataFrame(prices)
df['time'] = pd.to_datetime(df['time'], unit='s')
df.set_index('time', inplace=True)
df.head(10)
```
